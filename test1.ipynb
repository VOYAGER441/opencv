```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
# display image 
Image(filename='./image/test_1.png')
```

```python
# read image
ar=cv2.imread('./image/test_1.png',1)
```

```python
# the pic into the numpy array
print(ar)
```

```python
# the numpy array data 
print('image size is :',ar.shape)
# data type
print('image size is :',ar.dtype)
```

```python
# added color map
plt.imshow(ar,cmap="gray")
plt.axis("off")
```

```python
# reverse the color channel to bgr2rgb
ar_reverse=ar[:,:,::-1]
plt.imshow(ar_reverse)
plt.axis("off")
```

```python
# marge and split color channel
cat_pic_bgr=cv2.imread("./image/test_3.jpg",cv2.IMREAD_COLOR)
b,g,r=cv2.split(cat_pic_bgr)

# now subplot the images with diffrent color channel
plt.figure(figsize=[20,15])
# 1st subplot
plt.subplot(141)
plt.imshow(b,cmap="gray")
plt.title("Blue channel")
plt.axis("off")
# 2nd subplot
plt.subplot(142)
plt.imshow(g,cmap="gray")
plt.title("Green channel")
plt.axis("off")
# 3rd subplot
plt.subplot(143)
plt.imshow(r,cmap="gray")
plt.title("Red channel")
plt.axis("off")

# merge the 3 channel
cat_pic_marge=cv2.merge((b,g,r))

# now show the merge subplot with reverse color channel
plt.subplot(144)
plt.imshow(cat_pic_marge[:,:,::-1])
plt.title("Merge pic")
plt.axis("off")
```
